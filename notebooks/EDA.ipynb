```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
%matplotlib inline
# Loading dataset 
df = pd.read_csv('../data/score_updated.csv')
```

```python
df.head()
```

```python
# Dataset Information
print("Dataset Shape :", df.shape)
print("\nDataset Columns :", df.columns)
print("\nDataset Info :")
df.info()
```

```python
# Descriptive Statistics
print("Descriptive Statistics :")
print(df.describe())
```

```python
# Visualizing Distribution of study hours and marks
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(df['Hours'], bins=10 ,kde=True, color='blue')
plt.title('Distribution of Study Hours')

plt.subplot(1,2,2)
sns.histplot(df['Scores'], bins=10, kde=True, color="green")
plt.title("Distribution of Marks")

plt.show()
```

```python
# Scatter Plot Study Hours vs Marks
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Hours', y='Scores', data=df, color='purple')
plt.title('Study Hours vs Marks')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```

```python
# Correlation Analysis
corr = df.corr()
plt.figure(figsize=(5, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()

print("Correlation between Study Hours and Marks:",df["Hours"].corr(df["Scores"]))
```

```python
# Box Plots (check for outliers)
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.boxplot(y=df["Hours"], color='orange')
plt.title('Box Plot of Study Hours')

plt.subplot(1, 2, 2)
sns.boxplot(y=df["Scores"], color='teal')
plt.title('Box Plot of Marks')

plt.show()
```
